### Ещё раз проверю картинки, которые должны получиться

```python
model_new = {}
model_old = {}

for file in glob('outputs/model_new/*[0-9].csv'):
    table = pd.read_csv(file)
    model_new[table.emeas.mean()] = [ table.shape[0], table.total_soft_num.mean(), 20000 ]
    
for file in glob('outputs/model_old/*[0-9].csv'):
    table = pd.read_csv(file)
    model_old[table.emeas.mean()] = [ table.shape[0], table.total_soft_num.mean(), 20000 ]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model_new_df = pd.DataFrame.from_dict(model_new, orient='index').rename({0:'events', 1:'soft', 2:'total'}, axis=1)
model_old_df = pd.DataFrame.from_dict(model_old, orient='index').rename({0:'events', 1:'soft', 2:'total'}, axis=1)
```

```python
plt.figure(figsize=(13,7))
plt.errorbar(model_new_df.index, model_new_df.events/model_new_df.total, \
             yerr=np.sqrt(model_new_df.events)/model_new_df.total, fmt='o', label='new')
plt.errorbar(model_old_df.index, model_old_df.events/model_old_df.total, \
             yerr=np.sqrt(model_old_df.events)/model_old_df.total, fmt='o', label='old')
plt.grid(linestyle=":")
plt.legend(fontsize=16)
plt.title('Эффективность регистрации', fontsize=16);
```

```python
plt.figure(figsize=(13,7))
plt.errorbar(model_new_df.index, model_new_df.events/model_new_df.soft, \
             yerr=np.sqrt(model_new_df.events)/model_new_df.soft, fmt='o', label='new')
plt.errorbar(model_old_df.index, model_old_df.events/model_old_df.soft, \
             yerr=np.sqrt(model_old_df.events)/model_old_df.soft, fmt='o', label='old')
plt.grid(linestyle=":")
plt.legend()
plt.title('Эффективность регистрации');
```

```python
def pool_func(filename):
    a = HandleTree(filename, False)
    %time a.analyze()
    a.save_csv('outputs/2017/')
    return
```

```python
import multiprocessing as mp

# Step 1: Init multiprocessing.Pool()
pool = mp.Pool(3)

# Step 2: `pool.apply` the `howmany_within_range()`

results = pool.map(pool_func, tr_ph_2017)

# Step 3: Don't forget to close
pool.close()
```
