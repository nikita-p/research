```python
model_new = {}
model_old = {}

for file in glob('outputs/model_new/*.csv'):
    table = pd.read_csv(file)
    model_new[table.emeas.mean()] = [ table.shape[0], table.total_soft_num.mean(), 
                                         table.total_num.mean() ]
    
for file in glob('outputs/model_old/*.csv'):
    table = pd.read_csv(file)
    model_old[table.emeas.mean()] = [ table.shape[0], table.total_soft_num.mean(), 
                                         table.total_num.mean() ]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model_new_df = pd.DataFrame.from_dict(model_new, orient='index').rename({0:'events', 1:'soft', 2:'total'}, axis=1)
model_old_df = pd.DataFrame.from_dict(model_old, orient='index').rename({0:'events', 1:'soft', 2:'total'}, axis=1)
```
