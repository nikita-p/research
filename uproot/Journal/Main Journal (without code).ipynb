# Эффективность реконструкции $K_{S}$

## Объяснение

Эффективность реконструкции $K_{S}$ ($\varepsilon_{rec}$) - величина, показывающая отношение каонов, найденных процедурой поиска к общему числу каонов.
Существует разница между эффективностью реконструкции в эксперименте и моделировании, которую необходимо найти и учесть в измеренном сечении.
Эта поправка вычисляется следующим образом:

$$(1+\delta_{rec}) = \frac{\varepsilon_{rec}(exp)}{\varepsilon_{rec}(MC)}$$

Для определения эффективностей ($\varepsilon_{rec}$) в эксперименте и моделировании, необходимо чисто выделить события изучаемого процесса ($N_{total}$) и определить, какая доля этих событий прошла через стандартную процедуру поиска ($N_{passed}$):

$$\varepsilon_{rec} = \frac{N_{passed}}{N_{total}}$$

## Анализ

События изучаемого процесса определялись без использования блока $K_{S}$.

**Критерии поиска событий:**
1. Событие с двумя "хорошими" треками
1. Событие с кластером, коллинеарным к импульсу пары треков (расколлинеарность < 0.5) с энерговыделением > 100 МэВ
1. $| P_{\pi^+\pi^-} - \sqrt{E_{beam}^2 - m_{KS}^2} | < P_{cut}$, где $P_{cut} = 2 (0.0869  E_{beam} - 36.53)$ МэВ
1. $\frac{1}{2} | P_{\pi^+} + P_{\pi^-} | - \sqrt{ (E_{beam}/2)^2 - m_{\pi}^2} < 50$ МэВ (условие на распад на две одинаковые частицы из Ландау)
1. $\chi^2$ кинематической реконструкции (с неизвестным модулем импульса $K_S$) < 100
1. $|m_{\pi^+\pi^-} - m_{KS}| < 5$ МэВ, где $m_{KS}$ - среднемировая масса $K_S$

## Предположение о более сложной функции фита

Необходимо обосновать то, что одного гаусса для сигнала вполне достаточно.

Для этого можно использовать либо метод с построением истинных пионов, либо по-простому, найти $\chi^2$/ndf или FCN

### Тест MC Toy

Так как я фитирую свои распределения `Unbinned likelihood`, а в распределении по инвариантной массе есть большое количество нулевых бинов, то необходимо придумать способ оценки `goodness-of-fit`.

Суть метода MC Toy в том, чтобы сгенерировать большое количество распределений методом Монте-Карло, используя полученную в результате фита функцию, отфитировать их, получив таким образом распределение по значениям FCN, и понять, где лежит значение FCN моего фита в данном распределении.
Если фит плохой, то FCN будет лежать за пределами "гаусса" в распределении.
