### Как оценить поправку на неточное моделирование радиационного фотона:

Я подумал, что можно подойти к проблеме следующим способом:

пусть $\sigma_{r}$ - правильное (right) сечение с абсолютно точным моделированием

$$\sigma_{r} = \frac{n_0}{\varepsilon (1+\delta)} = \frac{n_0}{\frac{N_0}{N_0^t + N_\gamma^t} (1+\delta) } = \frac{n_0}{N_0/N_0^t} \frac{(1+N_\gamma^t/N_0^t)}{(1+\delta)}$$
где $n_0$ - количество событий "без" рад.фотона, $N_0$ - количество событий "без" рад.фотона, найденное в MC, $N_0^t$ - полное количество событий в MC "без" рад.фотонов, $N_\gamma^t$ - полное количество событий в MC с рад.фотонами

А если MC не такое точное, тогда я получу сечение $\sigma_{w}$ (wrong)

$$\sigma_{w} = \frac{n_0}{\varepsilon' (1+\delta)} = \frac{n_0}{\frac{{N'}_0}{ {N'}_0^t + {N'}_\gamma^t} (1+\delta) } = \frac{n_0}{{N'}_0/{N'}_0^t} \frac{(1+{N'}_\gamma^t/{N'}_0^t)}{(1+\delta)}$$

Из предыдущих формул получается, что

$$\sigma_{r} = \sigma_{w} \frac{{N'}_0/{N'}_0^t}{N_0/N_0^t} \frac{1+N_\gamma^t/N_0^t}{1+{N'}_\gamma^t/{N'}_0^t} $$

Где первый множитель - это поправка на эффективность реконструкции "упругого" процесса, а второй - поправка на события с рад. фотоном

Так как в любом случае невозможно отобрать события без радиационных фотонов, то здесь под этим подразумеваются мягкие фотоны ниже определённой энергии.

Можно найти эту поправку, построив распределение для доли излучённых мягких фотонов для моделирования и согласно радпоправке

![](outputs/images/10_part_mc.png)

Видно, что доля событий с мягким фотоном в MC систематически выше, чем согласно радпоправке.
Разделю одну величину на другую и отрисую поправку в зависимости от энергии
