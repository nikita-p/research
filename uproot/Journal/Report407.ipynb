#  Определение рад.поправок "методом СНД"

В данном методе

$$\sigma_{vis} = \int_{0}^{1} F(x,s) \sigma_{born}(s(1-x)) \varepsilon_{reg}(x,s) dx,$$

где $\sigma_{vis} = N/L$ - видимое сечение, $\sigma_{born}$ - борновское сечение, $\varepsilon_{reg}$ - эффективность регистрации в зависимости от энергии рад.фотона

так как $\sigma_{vis} = \sigma_{born} \varepsilon_{reg} (1 + \delta)$, то можно применять итерационный метод решения данного уравнения, подставив в нулевом приближении $\sigma_{born}^0 = \sigma_{vis}$, и, определяя $\varepsilon_{reg} (1 + \delta)$
