#### 3. Нужно оценить фон в сигнальной области

Видно, что сайд-бэнды не применить для оценки фона, т.к. в примыкающих к сигнальной области частях содержатся также события $e^+e^- \to K_SK_L$ только с более жёстким рад. фотоном.

Поэтому я снова взял мультиадронный генератор и посмотрел, есть ли события физического фона в отобранной области:

![](outputs/images/13_multi886.png)
![](outputs/images/13_multi924.png)

На данных картинках видно, что физический фон попадает в область отбора на уровне: 
1 событие $K^+$$K_S$$\pi^-$$\pi^0$ на 60 $K_L$$K_S$ на энергии 886 МэВ; 
1 событие $\pi^-$$\pi^+$$\pi^-$$\pi^+$$\pi^0$$\pi^0$ на 45 $K_L$$K_S$ на энергии 924 МэВ; 

#### 4. Ещё про фон

Из распределения по углу $\alpha$ между импульсом $K_S$ и направлением на ось пучков видно, что в неотобранной области лежит равномерный фон.
